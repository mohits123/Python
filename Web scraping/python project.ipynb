```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://finance.yahoo.com/most-active?offset=0&count=100"
page=requests.get(url)
page
```

```python
soup=BeautifulSoup(page.content)

```

```python
table=soup.find("table",class_="W(100%)")
```

```python
table=table.find("tbody")
```

```python
rows=table.find_all("tr",style="")
```

```python
market_info=[]
for i in rows:
    col=i.find_all("td")
    company_info=[c.text for c in col ]
    market_info.append(company_info)
print(market_info)
```

```python
symbol=[]
name=[]
price=[]
change=[]
per_change=[]
volume=[]
avg_vol=[]
market_cap=[]
pe_ratio=[]
week_52_range=[]
for i in market_info:
    symbol.append(i[0])
    name.append(i[1])
    price.append(i[2])
    change.append(i[3])
    per_change.append(i[4])
    volume.append(i[5])
    avg_vol.append(i[6])
    market_cap.append(i[7])
    pe_ratio.append(i[8])
    week_52_range.append(i[9])
```

```python
print(symbol)
```

```python
print(name)
```

```python
data={"Symbol":symbol,
      "Name":name,
      "Price":price,
      "Change":change,
      "Per_change":per_change,
      "Volume":volume,
      "Avg_vol":avg_vol,
      "Market_cap":market_cap,
      "Pe_ratio":pe_ratio,
      "Week_52_range":week_52_range}

   
      
     
      
      
      

      
      
```

```python
import pandas as pd
info=pd.DataFrame(data)
info
```

```python
info.to_csv("market.csv",index=False)
```
